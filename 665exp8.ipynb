```python
import math
import pandas as pd
from collections import Counter

documents = [
    "I love machine learning",
    "Machine learning is powerful",
    "I love AI"
    "there is one person"
]
processed_docs = [doc.lower().split() for doc in documents]
N = len(processed_docs)

vocab = sorted(set(word for doc in processed_docs for word in doc))
idf = {}
for word in vocab:
    df = sum(1 for doc in processed_docs if word in doc)
    idf[word] = math.log(N / df)

tfidf_matrix = []

for doc in processed_docs:
    tfidf_row = []
    word_freq = Counter(doc)
    total_words = len(doc)

    for word in vocab:
        tf = word_freq[word] / total_words
        tfidf = tf * idf[word]
        tfidf_row.append(round(tfidf, 4))

    tfidf_matrix.append(tfidf_row)

tfidf_df = pd.DataFrame(tfidf_matrix, columns=vocab)
tfidf_df.index = [f"Doc{i+1}" for i in range(N)]

print(tfidf_df)
```
