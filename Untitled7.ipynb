```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/sales_data_sample.csv", encoding="latin1")

# Preview data
print(df.head())

# Total Sales
total_sales = df["SALES"].sum()
print("\nTotal Sales:", total_sales)

# Average Sales
average_sales = df["SALES"].mean()
print("Average Sales:", average_sales)

# Sales by Product Line
print("\nSales by Product Line:")
print(df.groupby("PRODUCTLINE")["SALES"].sum())
```

```python
import pandas as pd

df = pd.read_csv("/content/sales_data_sample.csv", encoding="latin1")

# Year-wise sales
print("Year-wise Sales:")
print(df.groupby("YEAR_ID")["SALES"].sum())

# Deal size wise sales
print("\nDeal Size wise Sales:")
print(df.groupby("DEALSIZE")["SALES"].sum())
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/sales_data_sample.csv", encoding="latin1")

# Select input and output
X = df[["QUANTITYORDERED"]]   # Independent variable
Y = df["SALES"]              # Dependent variable

# Train the model
model = LinearRegression()
model.fit(X, Y)

# Predict sales for a new quantity
new_quantity = np.array([[40]])   # Example: 40 units
predicted_sales = model.predict(new_quantity)

print("Predicted Sales for Quantity = 40:", predicted_sales[0])
```
