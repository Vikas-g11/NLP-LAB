```python
!pip install spacy
!python -m spacy download en_core_web_sm

import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple is looking at buying a startup in India for $1 billion. It is a great opportunity."
doc = nlp(text)
```

```python
for token in doc:
    print(token.text)
```

```python
filtered = [token.text for token in doc if not token.is_stop]
print(filtered)
```

```python
!pip install nltk

from nltk.stem import PorterStemmer
ps = PorterStemmer()

for word in text.split():
    print(word, "->", ps.stem(word))
```

```python
for token in doc:
    print(token.text, token.pos_)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
